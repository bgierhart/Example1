```python
numbers = list() #this creates an empty list
for i in range(100000):
    x = random.randrange(10)
    numbers.append(x)
    #print(x, end=" ")
```

```python
plt.hist(numbers)
```

```python
gauss = list()
for i in range(1000):
    x = random.gauss(mu=16.0, sigma=2.5) #mu is mean, sigma is variance
    gauss.append(x)
```

```python
print(gauss)
```

```python
plt.hist(gauss)
```

```python
def print_numbers(*num): #*means that you can get any number of parameters in the function
    for n in num:
        print(n, end=" ")
```

```python
print_numbers(10,34,54,"RRK", 34, 89, 54)
```
